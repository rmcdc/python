# Reading and writing files in python

##Opening a file in python file = open("file path", "action"), action: r = read, w = write, a= append

```python
file = open("file.txt", "w")
```

```python
file
```

```python
# Writing to the file 
file.writelines(">Sequence1 \n")
file.writelines("ATTTCGCATT")
file.writelines("\n")
lst = [">Sequence2 \n", "ATGC"]
file.writelines(lst)
file.close()
```

```python
# opening file
file = open("file.txt", "r")
print(file.read())
file.close()
```

```python
# opening file and readline(), readlines()
file = open("file.txt", "r")
print(file.readlines())
file.close()
```
